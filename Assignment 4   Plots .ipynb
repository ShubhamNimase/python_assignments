# Plots on dataset

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df=pd.read_excel(r'E:\Superstore.xls')
```

```python
df.info()
```

```python
style.use("seaborn-dark")
plt.figure(figsize=(18,8))
cols=['r','violet','orange']
df.groupby('Category')['Sales'].sum().plot(kind='pie',
        colors=cols,startangle=90,shadow=True,textprops={'fontsize':20,'c':'black'},autopct='%0.2f%%')
plt.suptitle('Category wise sales',fontsize=25,color='b')
plt.show()

```

```python
style.use('seaborn-whitegrid')
plt.figure(figsize=(15,7))
df.groupby('Sub-Category')['Sales'].sum().plot(kind='bar',color='m',linewidth=0.5,fontsize=17)
plt.suptitle("Total sales per sub-category",fontsize=25,c='b')
plt.xlabel("Sub-Category",fontsize=20)
plt.ylabel('Total Sales',fontsize=20)
plt.show()
```

```python
style.use("seaborn-whitegrid")
plt.figure(figsize=(17,14))
df.groupby('State')['Sales'].mean().plot(kind='barh',color='b',linewidth=0.5,fontsize=16)
plt.suptitle("Statewise Sales",fontsize=20,color='black')
plt.ylabel('STATE',fontsize=18,c='k')
plt.xlabel('Sales',fontsize=18)
plt.show()
```

```python
df.groupby(['Category','Sub-Category'])['Sales'].sum().plot(kind='barh',color=['r','g','b'])
plt.show()
```

```python
style.use("seaborn-whitegrid")
plt.figure(figsize=(15,8))
df.groupby('Sub-Category')['Profit'].sum().plot(kind='barh',color=['r','g','b','m','orange','violet'],linewidth=0.5,fontsize=16)
plt.suptitle("Sub-Category wise Profit/Loss",fontsize=20,color='black')
plt.ylabel('Sub-category',fontsize=18,c='k')
plt.xlabel('Profit / Loss',fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(7,7))
df.groupby('Region')['Quantity'].sum().plot(kind='pie',shadow=True,
            textprops={'fontsize':15,'c':'black'},autopct='%0.2f%%',explode=(0,0,0.1,0),colors=['r','b','m','g'])
plt.suptitle("Region wise shipment quantity",fontsize=20)
plt.show()
```

```python
style.use('seaborn-whitegrid')
plt.figure(figsize=(7,7))
col=['r','g','b','m','y','orange','violet']
df.groupby(pd.to_datetime(df['Order Date'],format="%Y-%m-%d").dt.strftime('%a'))['Sales'].mean().plot(kind='pie',colors=col,
                            textprops={'fontsize':15,'c':'black'},autopct='%0.2f%%')
plt.suptitle('Day wise average Sales percentage',fontsize=20)
plt.show()
```

```python
style.use('seaborn-dark')
plt.figure(figsize=(7,7))
col=['r','g','b','m','y','orange','violet']
df.groupby(pd.to_datetime(df['Order Date'],format="%Y-%m-%d").dt.strftime('%b'))['Profit'].mean().plot(kind='pie',colors=col,
                            textprops={'fontsize':15,'c':'black'},autopct='%0.2f%%',explode=(0.2,0,0,0,0,0,0,0.2,0,0,0,0))
plt.suptitle('Month wise average profit percentage',fontsize=20)
plt.show()
```

```python
df.groupby('Product Name')['Quantity'].sum().sort_values().tail(15).plot(kind='barh',color='c')
plt.suptitle("Top 15 highest selling products",fontsize=15)
plt.xlabel('Quantity')
plt.show()
```

```python
fig=plt.figure(figsize=(12,10))
plt.subplot(221)
df.boxplot('Sales')
plt.subplot(222)
df.boxplot('Quantity')
plt.subplot(223)
df.boxplot('Discount')
plt.subplot(224)
df.boxplot('Profit')
plt.show()
```
