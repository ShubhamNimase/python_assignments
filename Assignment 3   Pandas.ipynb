```python
#5 month wise average of profit
df.groupby(pd.to_datetime(df['Order Date'],format="%Y-%m-%d").dt.strftime("%m"))['Profit'].mean()
```

```python
#7 Find sum of sales 
df["Sales"].sum()
```

```python
#8 Find state wise sum of sales.
df.groupby("State")['Sales'].sum()
```

```python
#9  Find month wise sum  of sales
df.groupby(pd.to_datetime(df['Order Date'],format="%Y-%m-%d").dt.month_name())['Sales'].sum()
```

```python
#10  Find country wise and city wise average profit
#city wise
df.groupby('City')['Profit'].mean()
```

```python
#country wise
df.groupby('Country')['Profit'].mean()
```

```python
#11 Calculate region wise average profit.
df.groupby('Region')['Profit'].mean()
```

```python
#12 Find the ship mode which is more profitable
df.groupby("Ship Mode")['Profit'].max()
```
