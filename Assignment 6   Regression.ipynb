# Regression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("boston.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df.iloc[:,:-1]
y=df['value']
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True)   #cmap='pink_r' ,'Set2'
plt.show()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(xtrain,ytrain)
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
y_pred=linreg.predict(xtest)
y_pred
```

```python
sns.regplot(ytest,y_pred)
```

```python
from sklearn import metrics
print('y-intercept :',linreg.intercept_)
print('beta coeeficients :',linreg.coef_)
print('Mean abs error MAE :',metrics.mean_absolute_error(ytest,y_pred))
print('Mean sq error MSE :',metrics.mean_squared_error(ytest,y_pred))
print('Root mean sq error PMSE :',np.sqrt(metrics.mean_squared_error(ytest,y_pred)))
print('r2 value :',metrics.r2_score(ytest,y_pred))
```

```python
import statsmodels.formula.api as smf
```

```python
df.rename(columns={'river"':'river'},inplace=True)
x.rename(columns={'river"':'river'},inplace=True)
```

```python
for i in x.columns: 
    print(i)
```

```python
model=smf.ols(formula='value~ crime_rate +proportion_res_land+proportion_business+river+NO+avg_rooms+proportion_units+distance+accessibility+property_tax+pupil_teacher_ratio+proportion_blacks+lower_status',data=df).fit()
model.summary()
```

## We will use forward selection method to determine significant variables dependng on their p-value

```python
xintercept=list(x.columns)
xintercept.insert(0,'Intercept')
#xintercept
```

```python
insig=[]
for i in list(zip(xintercept,model.pvalues)):
    if i[1]<=0.05:
        print("Reject Null:col  {:<20s} has strong relationship with response variable".format(i[0]))
    else:
        print("Accept Null:col  {:<20s} has weak relationship with response variable".format(i[0]))
        insig.append(i[0])
```

```python
print("Insignificant variables are :",insig)
```

## We will fit the model without these columns

```python
model2=smf.ols(formula='value~ crime_rate +proportion_res_land+river+NO+avg_rooms+distance+accessibility+property_tax+pupil_teacher_ratio+proportion_blacks+lower_status',data=df).fit()
model2.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(x.values, i)
                          for i in range(len(x.columns))]
  
print(vif_data)
```
